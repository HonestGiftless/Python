<h1 align="center">Интерфейс операционной системы</h1>

* Если нужно просто прочитать или записать файл, лучше использовать встроенную функцию open()
* Для различных манипуляций с путями удобнее использовать os.path или pathlib
* Если нужно прочитать все строки в файлах, указанных в командной строке - fileinput
* Для создания временных файлов и каталогов - tempfile
* Для операций с файлами и каталогами (копирование, перемещение, создание, удаление) - shutil
